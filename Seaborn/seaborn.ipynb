```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

```python
# Dados de Covid
covid_df = pd.read_csv("https://raw.githubusercontent.com/covid19br/covid19br.github.io/master/dados/SRAGs-tabela-last-updated_revised2.csv")
```

```python
covid_df.head()
```

```python
plt.bar(x=covid_df['Data'], height=covid_df['27/04/20'])
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=covid_df, x='Data', y='27/04/20')

plt.xticks(rotation=90)
```

```python
dados = covid_df[["Data", "08/04/20", "20/04/20", "27/04/20"]]
```

```python
dados.head()
```

```python
dados = dados.melt('Data', var_name="Reporte", value_name="Obitos")
```

```python
dados.head()
```

```python
plt.figure(figsize=(15,5))

sns.barplot(data=dados, x='Data', y='Obitos', hue='Reporte')
plt.xticks(rotation=90)
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(14,8))
plt.ylim(0, 300)
plt.xlim(0, 54)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)

def plot_obitos(idx):
    colunas = covid_df.columns[::-1]
    dados = pd.DataFrame(covid_df[colunas[:idx]])
    dados['Data'] = covid_df['Data']
    
    # plot do passado
    dados = dados.melt('Data', var_name='Reporte', value_name='Obitos')
    sns.lineplot(data=dados, x='Data', y='Obitos', hue='Reporte', palette='g', ax=ax)
    
    # plot do presente
    sns.lineplot(data=dados, x='Data', y=colunas[idx], hue='Reporte', palette='r', ax=ax)
    
plot_obitos(1)
```
