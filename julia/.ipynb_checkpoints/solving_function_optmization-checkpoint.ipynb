```julia
function func(x)
    y=x^2
end 
```

```julia
# Calcula a funcao derivada (gradiente)
function gradient(func, x)
    # App. Derivative
    h = 1e-8
    y = func(x)
    y_gd_approx = (func(x+h) - y) / h
    return y_gd_approx, y # In Julia one can use return or simply by default the  last expression value
end
```

```julia
function gradient_line_eq_(x, m, x_point, y_point)
    c = y_point - m*x_point
    y = m*x + c
    return y, c
end
```

```julia
function solving_the_linear_programming(m)
    #https://www.softcover.io/read/7b8eb7d0/juliabook2/linear
    JuMP.optimize!(m)    
    x = JuMP.value(x_var)
    y = JuMP.value(y_var)
    return x, y
end
```
