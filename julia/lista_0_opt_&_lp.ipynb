Resolvendo o problema como uma otimização. É buscado então, a minimização do erro entre o resultado do modelo e os valores observados.

Equação da Regressão:

$
y_i (Contratos) = \beta_0 + \beta_1 x_i(Preço Spot) + \epsilon_i 
$

Com isto, pode-se rescrever a equação como

$
\epsilon_i  = y_i - (\beta_0 + \beta_1 x_i) 
$

E então busca-se minimizar

$
argmin \sum_{i=1}^{n}\epsilon_i^2
$

```julia
function linear_reg_solver(x, y)
    n = length(x)
    m = Model(HiGHS.Optimizer)
    @variable(m, b0)
    @variable(m, b1)
    @variable(m, residuals[1:n])
    @constraint(m, [i in 1:n], residuals[i] == y[i] - b0 - b1*x[i])
    @objective(m, Min, sum(residuals.^2))
    optimize!(m)
    return m, b0, b1
end
```

```julia
m, b0, b1 = linear_reg_solver(data.Spot, data.Contratos)
solution_summary(m)
value(b0), value(b1)
```

Resolvendo com o modelo regressão quantílica (para τ=0.10, 0.25, 0.50, 0.75, 0.90)

Utilizando o modelo de regressão logistica busca-se, de forma semelhante à regressão linear, encontrar os coeficientes $\beta$ do modelo.
Porém a função que é minimizada neste caso é:

$
argmin \sum_{i=1}^{n} \lambda |e_i'| + (1 - \lambda) |e_i''|
$

Sendo $e'$ e $e''$ dados por:

$
e' + e'' = y_i - (\beta_0 + \beta_1 x_i) 
$

```julia
function quant_reg_solver(x:: Vector{Float64}, y:: Vector{Float64}; lamb::Float64=0.5)
    n = length(x)
    m = Model(HiGHS.Optimizer)
    @variable(m, b0)
    @variable(m, b1)
    @variable(m, residuals[1:n])
    @variable(m, residuals_upper[1:n] >=0)
    @variable(m, residuals_lower[1:n] >=0)
    # @constraint(m, [i in 1:n], residuals[i] == y[i] - b0 - b1*x[i])
    @constraint(m, [i in 1:n], residuals_upper[i] >= y[i] - b0 - b1*x[i])
    @constraint(m, [i in 1:n], residuals_lower[i] >= b0 + b1*x[i] - y[i])

    # @constraint(m, [i in 1:n], residuals_upper[i] - residuals_lower[i] == residuals[i])
    
    @objective(m, Min, sum((lamb)*residuals_upper[i] + (1-lamb)*residuals_lower[i] for i=1:n))
    optimize!(m)
    return m, b0, b1, residuals_upper, residuals_lower
end
```

```julia
m, b0, b1, residuals_upper, residuals_lower = quant_reg_solver(data.Spot, data.Contratos, lamb=0.1)
value(b0), value(b1)
m, b0, b1, residuals_upper, residuals_lower = quant_reg_solver(data.Spot, data.Contratos, lamb=0.25)
value(b0), value(b1)
m, b0, b1, residuals_upper, residuals_lower = quant_reg_solver(data.Spot, data.Contratos, lamb=0.50)
value(b0), value(b1)
m, b0, b1, residuals_upper, residuals_lower = quant_reg_solver(data.Spot, data.Contratos, lamb=0.75)
value(b0), value(b1)
m, b0, b1, residuals_upper, residuals_lower = quant_reg_solver(data.Spot, data.Contratos, lamb=0.75)
value(b0), value(b1)
```

### Ex. 2. Hodrick-Prescott Filter

O filtro de HP pode ser escrito da forma:

$
hp = \sum (y_t - \tau_t)^2 + \lambda \sum{} ((\tau_{t+1} - \tau_t) - (\tau_t - \tau_{t-1}))^2
$

Reescrevendo a equação como um problema de programação linear, as parcelas da equação do filtro podem ser abertas de forma modular. Tem-se assim que:

$
e_{upper} = y_t - \tau_t
$

$
e_{lower} = \tau_t - y_t
$

$
der_{upper} = (\tau_{t+1} - \tau_t) - (\tau_t - \tau_{t-1})
$

$
der_{lower} = (\tau_t - \tau_{t-1}) - (\tau_{t+1} - \tau_t)
$

Desta forma a equação de minimização será então,

$
\sum (e_{upper} + e_{lower}) + \lambda \sum (der_{upper} + der_{lower})
$

```julia
function hp_abs_solver(y:: Vector{Float64}; tau:: Int64=1)
    n = length(y)
    m = Model(HiGHS.Optimizer)
    @variable(m, t[1:n])
    @variable(m, residuals_lower[1:n] >=0)
    @variable(m, residuals_upper[1:n] >=0)
    @variable(m, der_upper[1:n] >=0)
    @variable(m, der_down[1:n] >=0)
    
    # @variable(m, residuals[1:n])

    # @constraint(m, [i in 1:n], residuals[i] == y[i] - b0 - b1*x[i])
    @constraint(m, [i in 1:n], residuals_upper[i] >= y[i] - t[i])
    @constraint(m, [i in 1:n], residuals_lower[i] >= t[i] - y[i])

    @constraint(m, [i in 2:n-1], der_upper[i] >= (t[i+1] - t[i]) - (t[i] - t[i-1]))
    @constraint(m, [i in 2:n-1], der_down[i] >= (t[i] - t[i-1]) - (t[i+1] - t[i]))

    @objective(m, Min, sum(residuals_upper[i] + residuals_lower[i] for i=1:n) + 
        tau*( sum(der_upper[i] + der_down[i] for i=2:n-1) )
    )
    optimize!(m)
    return t, residuals_upper, residuals_lower
end
```

Aplicando o filtro HP com o termo de minimização quadrático,

```julia
function hp_quadratic_solver(y:: Vector{Float64}; tau:: Int64=1)
    n = length(y)
    # m = Model(Mosek.Optimizer)
    m = Model(Ipopt.Optimizer)
    # m = Model(HiGHS.Optimizer)
    @variable(m, t[1:n])
    
    @variable(m, residuals[1:n])
    @variable(m, der[1:n])

    @constraint(m, [i in 1:n], residuals[i] == y[i] - t[i])
    @constraint(m, [i in 2:n-1], der[i] >= (t[i+1] - t[i]) - (t[i] - t[i-1]))

    @objective(m, Min, sum(residuals[i]^2 for i=1:n) + tau*sum(der[i]^2 for i=2:n-1))
    optimize!(m)
    return t, residuals
end
```
